```python
from joblib import dump, load
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
def score_function(y_truth, y_pred, range_threshold=2):
    # Calculate the absolute difference between y_pred and y_truth
    diff = np.abs(y_pred.astype(float) - y_truth.astype(float))
    
    # Create a mask for values within the specified range_threshold
    mask = diff <= range_threshold
    
    # Calculate the score by counting the positive predictions within the range threshold
    score = np.mean(mask)
    
    return score
```

<font size='20'>Logistic Regression</font>

```python
mod = load('./model/Logistic_Regression_mmse.joblib')
mod2 = load('./model/Logistic_Regression_dx.joblib')

X_test, y_test = np.load('./test_data.npz')['X_test'], np.load('./test_data.npz')['y_test']
y_test_mmse = y_test[:,0]
y_test_dx = y_test[:,1]

print(score_function(y_test_mmse, mod.predict(X_test)))
print(mod2.score(X_test, y_test_dx))

cm = confusion_matrix(y_test_dx, mod2.predict(X_test), labels=mod2.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=mod2.classes_)
disp.plot()
```
